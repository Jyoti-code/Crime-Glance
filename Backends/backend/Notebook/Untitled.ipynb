```python
df.insert(20,"SafeMeasure",0)
```

```python
for i in range(len(df)):
  if df.Crime_rate[i]>150:
    df.SafeMeasure[i]=0
  elif df.Crime_rate[i]>100 and df.Crime_rate[i]<150:
    df.SafeMeasure[i]=1
  else:
    df.SafeMeasure[i]=2
```

```python
x = df[['GDPP','TotalPop','MALE_LIT']]
y= df['SafeMeasure']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```
